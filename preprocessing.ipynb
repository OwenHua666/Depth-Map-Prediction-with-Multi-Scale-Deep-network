```python
import torch
import numpy as np
from scipy.io import loadmat
from scipy.io import savemat
import scipy
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
```

```python
from skimage.transform import resize
```

```python
# load dataset
images=loadmat('./data/images.mat')
depths=loadmat('./data/depths.mat')
image_data=images['images']
print(image_data.shape)
print(type(image_data))
image_depths=depths['depths']
print(image_depths.shape)
print(type(image_depths))
```

```python
sample = {'image': image_data, 'depth': image_depths}
input_height = 228
input_width = 304
```

```python
# resize image
resized_image=np.zeros((228,304,3,1449))
resized_depth=np.zeros((228,304,1449))
for i in range(0,1449):
    resized_image[:,:,:,i]=scipy.misc.imresize(np.squeeze(sample['image'][:,:,:,i]),(228,304,3))
    resized_depth[:,:,i]=scipy.misc.imresize(np.squeeze(sample['depth'][:,:,i]),(228,304))
```

```python
# resize image
resized_image=np.zeros((228,304,3,1449))
resized_depth=np.zeros((228,304,1449))
for i in range(0,1449):
    resized_image[:,:,:,i]=resize(np.squeeze(sample['image'][:,:,:,i]),(228,304,3))
    resized_depth[:,:,i]=scipy.misc.imresize(np.squeeze(sample['depth'][:,:,i]),(228,304))
```

```python
# visualize original and resized images and depths
plt.subplot(121)
plt.imshow(sample['image'][:,:,:,0])
plt.title('original image')
plt.subplot(122)
plt.imshow(resized_image[:,:,:,0])
plt.title('resized image')
plt.show()

plt.subplot(121)
plt.imshow(sample['depth'][:,:,0])
plt.title('original depth')
plt.subplot(122)
plt.imshow(resized_depth[:,:,0])
plt.title('resized depth')
plt.show()
```

```python
resized_data={'image':resized_image,'depth':resized_depth}
```

```python
savemat('resized_data',resized_data)
```

```python
#split data set 90-10
np.random.seed(0)
order = np.random.permutation(1449) #shuffle
images_shuffled = resized_image[:,:,:, order]
depths_shuffled = resized_depth[:, :,order]
images_train=images_shuffled[:,:,:,0:1159]
images_test=images_shuffled[:,:,:,1159:]
depths_train=depths_shuffled[:,:,0:1159]
depths_test=depths_shuffled[:,:,1159:]
```

```python
# visualize training and test images and dep
plt.subplot(121)
plt.imshow(images_train[:,:,:,0])
plt.title('training image')
plt.subplot(122)
plt.imshow(depths_train[:,:,0])
plt.title('training depth')
plt.show()

plt.subplot(121)
plt.imshow(images_test[:,:,:,0])
plt.title('test image')
plt.subplot(122)
plt.imshow(depths_test[:,:,0])
plt.title('test depth')
plt.show()
```

```python
tensor_image_train = torch.from_numpy(images_train)
tensor_image_test = torch.from_numpy(images_test)
tensor_depths_train = torch.from_numpy(depths_train)
tensor_depths_test = torch.from_numpy(images_test)
print(
type(tensor_image_train),
type(tensor_image_test),
type(tensor_depths_train),
type(tensor_depths_test),)
```
